```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")


print(netflix_df["genre"])
netflix_df_movies = netflix_df[netflix_df["genre"] != "International TV"]
movies_1990s = netflix_df_movies[(netflix_df_movies["release_year"]>=1990) &  (netflix_df_movies["release_year"]<=1999)]
print(movies_1990s["release_year"].min())
plt.hist(movies_1990s["duration"], bins=30)
plt.show()
plt.clf()
duration = 94 # rounded
action_movies = movies_1990s[movies_1990s["genre"] == "Action"]
short_movie_count = 0
for label, row in action_movies.iterrows() :
    if row["duration"] <90:
        short_movie_count = short_movie_count + 1

print(short_movie_count)
```
